# Caption Generator

A **Caption Generator** is an AI model that analyzes an image and generates a natural-language description of its contents.

**Example:**
- **Input:** Image of a dog playing with a ball.
- **Output:** *"A dog is playing with a ball in the park."*

**Uses:**
- Image description
- Social media captions
- Accessibility for visually impaired users

---

# OCR (Optical Character Recognition)

**OCR** is a technology that extracts and converts text from images or scanned documents into editable, machine-readable text.

**Example:**
- **Input:** Image containing `WELCOME`
- **Output:** `WELCOME`

**Uses:**
- Document digitization
- Receipt and invoice processing
- Text extraction from images

---

## Difference

| Caption Generator | OCR |
|-------------------|-----|
| Describes an image | Extracts text from an image |
| Output is a sentence | Output is the exact text |
| Understands the scene | Reads only the text |

```python
from google.colab import userdata
from openai import OpenAI
```

```python
api_key = userdata.get('api')
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key,
)
```

```python
from google.colab import files
import cv2
from google.colab.patches import cv2_imshow
uploaded = files.upload()
for filename in uploaded.keys():
  img = cv2.imread(filename)
  print(img)
  cv2_imshow(img)
```

```python
import base64

for filename in uploaded.keys():
  with open(filename, "rb") as image_file:  # rb= read binary
    image_bytes = image_file.read()

    base64_string = base64.b64encode(image_bytes).decode("utf-8")  # binary to string conversation
    print(base64_string)
    # print(base64_string[:200])
    print(f"\nLength: {len(base64_string)}")
```

```python
response = client.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "Describe this image."
                },
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base64_string}"  #to specify the backend, we are using image
                    }
                }
            ]
        }
    ],
    max_tokens=500
)
reply = response.choices[0].message.content
reply
```

```python
response = client.chat.completions.create(
    model="nvidia/nemotron-3-nano-omni-30b-a3b-reasoning:free",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "Describe this image."
                },
                {
                    "type": "image_url",
                    "image_url": {
                        "url": "https://i.pinimg.com/736x/46/78/c2/4678c25fbdf394b37e7e313c23ba4d1e.jpg"
                    }
                }
            ]
        }
    ],
    max_tokens=500
)
reply = response.choices[0].message.content
reply
```

# Quantization

**Quantization** is the process of reducing the precision of a machine learning model's weights and activations (e.g., from **32-bit floating point (FP32)** to **8-bit integers (INT8)**) to make the model smaller and faster.

**Benefits:**
- Reduces model size
- Speeds up inference
- Lowers memory usage
- Enables deployment on resource-constrained devices
