# **Topics covered:** autograd basics (gradients, retain_grad), building a neural network with nn.Module, bias, and activation functions (ReLU, Leaky ReLU).

```python
import torch
import torch.nn as nn
```

```python
x = torch.tensor(3.0 , requires_grad=True)

a = x+4       # a is non-leaf (result of an op)
b = a*5
a.retain_grad()   # needed so a.grad isn't dropped after backward

b.backward()

print("x.grad = ", x.grad)
print("a.grad = ", a.grad)
```

```python
x = torch.tensor(3.0 , requires_grad=True)

a = x**2
b = a**2
a.retain_grad()

b.backward()

print("x.grad = ", x.grad)   # db/dx
print("a.grad = ", a.grad)   # db/da
```

```python
x = torch.tensor(3.0 , requires_grad=True)

y = x**2

g = torch.autograd.grad(outputs= y , inputs= x)

print(g)
print(x.grad)
```

```python
x = torch.tensor(2.0 , requires_grad=True)

y = (x**5) + 3*x

g = torch.autograd.grad(outputs= y , inputs= x)

print(g)
print(x.grad)
```

# NN--NEURAL NETWORKS
How to define a model in PyTorch using nn.Module.

```python
class MyModel(nn.Module):

  def __init__(self):
    super().__init__()

  def forward(self, x):
    return x+3
```

```python
myModel = MyModel()
myModel
```

```python
class MyModel(nn.Module):

  def __init__(self):
    super().__init__()

    self.linear = nn.Linear(2, 1)

  def forward(self, x):
    return self.linear(x)
```

```python
model = MyModel()

print(model)
```

RELU--Rectified Linear Unit


*   positive values remain as it is
*   negative values changes to 0(zero)




```python
relu = nn.ReLU()
```

```python
x = torch.tensor([-3. , -4. , 2. , 6.])

y = relu(x)

print(y)
```

```python
x = torch.tensor([(-3.*-3. ), -4. , 2. , 6. , 0 , -0, -0. , 0.])

y = relu(x)
y = y + 0.0   # -0. and 0. both clamp to 0. under ReLU
print(y)
```

# Leaky Relu
Like ReLU, but lets a small negative slope through instead of zeroing it out.

```python
x = torch.tensor([-5., -2., 0., 3., 7.])

leaky_relu = nn.LeakyReLU()  #()--slope---by default=0.01 , we can also change it

y = leaky_relu(x)

print(y)
```
