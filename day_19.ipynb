```python
from openai import OpenAI
from google.colab import userdata
```

```python
api_key=userdata.get('GEMINI_API')
```

```python
client = OpenAI(
    base_url="https://generativelanguage.googleapis.com/v1beta",
    api_key=api_key
)
```

Temperature--

*   0 = model should not do its creativity
*   0.5-0.7 = model should use its medium creativity
*   1-2 = model can use its high creativity[link text](https://)

```python
response = client.chat.completions.create(
    model = "gemini-2.5-flash-lite",
    messages = [{'role': 'user', 'content': 'hi,this is test'}], max_tokens=5
)
response.choices[0].message.content
```

```python
system_promp = {'role':'system', 'content': 'You are a helpful agent who helps users to select their next tourist place, You will ask their country and budget'}
messages = [system_promp]
user = input('Enter your prompt: ')
messages.append({'role': 'user', 'content': user})
```

```python
messages
```

```python
response = client.chat.completions.create(
    model = "gemini-2.5-flash-lite",
    messages = messages
)
```

```python
messages.append({'role': 'assistant', 'content': response.choices[0].message.content})
response.choices[0].message.content
```
