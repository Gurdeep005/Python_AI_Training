# NN-
Defining a simple neural network and running a forward pass on it.

```python
class StudentModel(nn.Module):

  def __init__(self):
    super().__init__()

    self.layer1 = nn.Linear(2,4)
    self.relu = nn.ReLU()
    self.layer2 = nn.Linear(4,1)

  def forward(self, x):

    x = self.layer1(x)
    x = self.relu(x)
    x = self.layer2(x)

    return x
```

```python
model = StudentModel()
```

MSELoss (Mean Squared Error) — Common for predicting continuous numbers (like exam scores or house prices).
CrossEntropyLoss — Common for classification problems (like predicting whether an email is spam or not).

```python
import torch
import torch.nn as nn

class StudentModel(nn.Module):

    def __init__(self):
        super().__init__()

        self.layer1 = nn.Linear(2, 4)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(4, 1)

    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        return x


model = StudentModel()

student = torch.tensor([[6.0, 80.0]])
actual_score = torch.tensor([[90.0]])

loss_fn = nn.MSELoss()

prediction = model(student)

loss = loss_fn(prediction, actual_score)

print("Prediction:", prediction)
print("Loss:", loss)
print(model.layer1.weight.grad)
loss.backward()
print(model.layer1.weight.grad)
```

```python
print(model.layer1.bias.grad)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

class StudentModel(nn.Module):

    def __init__(self):
        super().__init__()

        self.layer1 = nn.Linear(2, 4)

        self.relu = nn.ReLU()

        self.layer2 = nn.Linear(4, 1)

    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        return x


model = StudentModel()


student = torch.tensor([[6.0, 80.0]])
actual_score = torch.tensor([[90.0]])


loss_fn = nn.MSELoss()


optimizer = optim.SGD(model.parameters(), lr=0.001)


prediction = model(student)


loss = loss_fn(prediction, actual_score)


optimizer.zero_grad()


loss.backward()


optimizer.step()


print("Prediction:", prediction)
print("Loss:", loss)
```

```python
model.parameters()
```

```python
for epoch in range(100000):   # training loop: predict -> compute loss -> backprop -> update weights
    prediction = model(student)

    loss = loss_fn(prediction, actual_score)

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()
```

# Practice task: build and train a model to predict a car's mileage.

```python
import torch
import torch.nn as nn
import torch.optim as optim

class Cars(nn.Module):

    def __init__(self):
        super().__init__()

        self.layer1 = nn.Linear(2, 4)

        self.relu = nn.ReLU()

        self.layer2 = nn.Linear(4, 3)

        self.gelu = nn.GELU()

        self.layer3 = nn.Linear(3, 1)

    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        x = self.gelu(x)
        x = self.layer3(x)
        return x
```

```python
model = Cars()
```

```python
car1 = torch.tensor([5., 120.])
```

```python
loss_fn = nn.MSELoss()
```

```python
optimizer = optim.SGD(model.parameters(), lr=0.001)
actual_score = torch.tensor([25.0])

for epoch in range(100000):

    prediction = model(car1)

    loss = loss_fn(prediction, actual_score)

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()
```

```python
prediction = model(car1)
```
