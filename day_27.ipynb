```python
arr = np.array([1,2,3])
```

```python
tens = torch.tensor(arr)
```

```python
print(arr)
print(tens)
```

when we make changes to an array, it will also affect tensor(only when both arr and tens on CPU), but when we run arr on CPU and tens run on GPU , it will not effect.

```python
arr[1]=9
```

```python
print(arr)
print(tens)
```

We can specify TENSOR what to use, whether- CPU or GPU

```python
torch.tensor(1,  device="cuda")
```

```python
a = torch.tensor(1,  device='cuda')
b = torch.tensor(4,  device='cuda')
```

```python
a+b
```
